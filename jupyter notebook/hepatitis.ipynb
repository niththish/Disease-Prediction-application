```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("hepatit-c-data.csv")
dataset.head()
```

```python
dataset.describe()
```

```python
dataset.isnull().any()
```

```python
dataset["ALB"] = dataset["ALB"].fillna(dataset["ALB"].mean())
dataset["ALP"] = dataset["ALP"].fillna(dataset["ALP"].mean())
dataset["ALT"] = dataset["ALT"].fillna(dataset["ALT"].mean())
dataset["CHOL"] = dataset["CHOL"].fillna(dataset["CHOL"].mean())
dataset["PROT"] = dataset["PROT"].fillna(dataset["PROT"].mean())
```

```python
dataset.isnull().any()
```

```python
dataset = dataset.drop("Unnamed: 0",axis=1)
```

```python
dataset["Category"].unique()
```

```python
dataset["Category"] = dataset["Category"].replace("0=Blood Donor",0)
dataset["Category"] = dataset["Category"].replace("1=Hepatitis",1)
dataset["Category"] = dataset["Category"].replace("2=Fibrosis",2)
dataset["Category"] = dataset["Category"].replace("3=Cirrhosis",3)
dataset["Category"] = dataset["Category"].replace("0s=suspect Blood Donor",4)

dataset["Sex"] = dataset["Sex"].replace("m",0)
dataset["Sex"] = dataset["Sex"].replace("f",1)
```

```python
dataset.dtypes
```

```python
print("----------------Mean---------------\n",dataset.mean(),"\n---------------------------------------")
print("----------------Median---------------\n",dataset.median(),"\n---------------------------------------")
print("----------------Standard deviation---------------\n",dataset.std(),"\n---------------------------------------")
```

```python
dataset.groupby('Category').mean()
```

```python
X = dataset.drop("Category",axis=1)
Y = dataset["Category"]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
```
